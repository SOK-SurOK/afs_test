## Импортируем нужные модули и загружаем csv

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
big_data = pd.read_csv("test.csv", delimiter=',')
big_data = big_data.rename(columns={
    "Month_of_2020": "month_of_2020", "Game name": "game_name", "Payment System": "payment_system",
    "User country": "user_country", "Incomplete attempts": "incomplete_attempts",
    "Completed payments": "completed_payments", "AFS reject": "AFS_reject",
    "PS declines": "PS_declines", "Refunds": "refunds"
})
big_data
```

## Разделяй и властвуй
поделим таблицу на подтаблицы: дано только 2 игры и 5 месяцев - воспользуемся этим

```python
game1 = big_data[big_data["game_name"] == " Funny squirrel"].drop("game_name", 1)
game2 = big_data[big_data["game_name"] == " Funny squirrel in Space"].drop("game_name", 1)
game1_month_list = []
game2_month_list = []
for i in range(5):
    game1_month_list.append(game1[game1["month_of_2020"] == i+1 ].drop("month_of_2020", 1))
    game2_month_list.append(game2[game2["month_of_2020"] == i+1 ].drop("month_of_2020", 1))
print("Успешно")
```

### Сначала я хотел сгруппировать по типу:

    game2_month_list[i].groupby(["payment_system", "user_country"]).mean()

Но как оказалось эти два поля для подтаблиц являются почти PRIMARY KEY  
Поэтому я решил группировать по каждому из этих полей:

```python
plt.rcParams["figure.figsize"] = (20,10)
for i in range(len(game1_month_list)):
    view = game1_month_list[i].groupby("user_country", as_index = False).sum().sort_values("user_country")
    view.plot(x="user_country", kind="bar", fontsize=6, title="месяца "+str(i+1)+"-"+ str(i+2), y=[
        "incomplete_attempts", "completed_payments", "AFS_reject", "PS_declines", "refunds"
    ])
```

### Но как видно - ничего не понятно
(для второй игры и другой группировки выполняется аналогично, но чтобы не загромождать мусором - не стал включать)
### Воспользуемся разницей между ближайшими месяцами и успешные транзакции уберем, так как слишком большой вклад вносят в размер таблицы (для второй игры и группировки по payment_system проделывается тоже самое, но я дабы не загромождать не стал этого делать):

```python
plt.rcParams["figure.figsize"] = (20,10)
for i in range(len(game1_month_list) - 1):
    view0 = game1_month_list[i].groupby("user_country", as_index = False).sum().sort_values("user_country")
    view = view0.drop("user_country", 1)
    view_next = game1_month_list[i + 1].groupby("user_country", as_index = False).sum().sort_values("user_country")
    view_next = view_next.drop("user_country", 1)
    view_result = view_next - view
    view_result.insert(0, "user_country", view0["user_country"], True)
    view_result.plot(x="user_country", kind="bar", fontsize=6, title="месяца "+str(i+1)+"-"+ str(i+2), y=[
        "incomplete_attempts", "AFS_reject", "PS_declines", "refunds"
    ])
```

### Для большей наглядности поделим на максимальное отклонение (теперь можно вернуть "completed_payments")

```python
plt.rcParams["figure.figsize"] = (20,10)
for i in range(len(game1_month_list) - 1):
    view0 = game1_month_list[i].groupby("user_country", as_index = False).sum().sort_values("user_country")
    view = view0.drop("user_country", 1)
    view_next = game1_month_list[i + 1].groupby("user_country", as_index = False).sum().sort_values("user_country")
    view_next = view_next.drop("user_country", 1)
    view_result0 = view_next - view
    
    view_result = view_result0.abs()
    view_result /= view_result.max()
    view_result *= (view_result0 / view_result0.abs())
    
    view_result.insert(0, "user_country", view0["user_country"], True)
    view_result.plot(x="user_country", kind="bar", fontsize=6, title="месяца "+str(i+1)+"-"+ str(i+2), y=[
        "incomplete_attempts", "completed_payments", "AFS_reject", "PS_declines", "refunds"
    ])
```
